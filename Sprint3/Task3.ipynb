#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
using System;
using System.Collections.Generic;

// Базовый класс
public class Book
{
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    
    // Новые атрибуты
    public string ISBN { get; set; }
    public int PageCount { get; set; }
    public string Publisher { get; set; }
    public bool IsAvailable { get; protected set; } = true;

    public Book(string title, string author, int yearOfPublication)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
    }

    // Перегруженный конструктор
    public Book(string title, string author, int yearOfPublication, 
                string isbn, int pageCount, string publisher) 
        : this(title, author, yearOfPublication)
    {
        ISBN = isbn;
        PageCount = pageCount;
        Publisher = publisher;
    }

    public virtual string GetInfo()
    {
        return $"Название: {Title}, Автор: {Author}, Год: {YearOfPublication}, Страниц: {PageCount}";
    }

    public virtual void Read()
    {
        Console.WriteLine($"Чтение книги: {Title}");
        MarkAsRead();
    }

    // Перегруженный метод Read
    public virtual void Read(int pages)
    {
        Console.WriteLine($"Чтение {pages} страниц книги: {Title}");
        if (pages > 0) MarkAsRead();
    }

    public virtual void Borrow()
    {
        if (IsAvailable)
        {
            Console.WriteLine($"Выдача книги: {Title}");
            IsAvailable = false;
        }
        else
        {
            Console.WriteLine($"Книга {Title} уже выдана");
        }
    }

    // Новые методы
    public virtual void Return()
    {
        IsAvailable = true;
        Console.WriteLine($"Книга {Title} возвращена");
    }

    public virtual void Review()
    {
        Console.WriteLine($"Просмотр обзора книги: {Title}");
    }

    protected virtual void MarkAsRead()
    {
        Console.WriteLine($"Книга {Title} отмечена как прочитанная");
    }
}

public class Textbook : Book
{
    public string Subject { get; set; }
    public int Edition { get; set; }
    public string GradeLevel { get; set; }
    public bool HasExercises { get; set; }

    public Textbook(string title, string author, int yearOfPublication, string subject) 
        : base(title, author, yearOfPublication)
    {
        Subject = subject;
    }

    // Перегруженный конструктор
    public Textbook(string title, string author, int yearOfPublication, 
                   string subject, int edition, string gradeLevel, bool hasExercises)
        : base(title, author, yearOfPublication)
    {
        Subject = subject;
        Edition = edition;
        GradeLevel = gradeLevel;
        HasExercises = hasExercises;
    }

    public override void Read()
    {
        Console.WriteLine($"Изучение учебника по предмету '{Subject}': {Title}");
        MarkAsRead();
    }

    // Переопределение с перегрузкой
    public override void Read(int pages)
    {
        Console.WriteLine($"Изучение {pages} страниц учебника '{Subject}': {Title}");
        if (pages > 0 && HasExercises)
        {
            DoExercises();
        }
    }

    // Новые методы
    public void DoExercises()
    {
        Console.WriteLine($"Выполнение упражнений из учебника: {Title}");
    }

    public void UpdateEdition(int newEdition)
    {
        Edition = newEdition;
        Console.WriteLine($"Учебник {Title} обновлен до {newEdition} издания");
    }

    protected override void MarkAsRead()
    {
        Console.WriteLine($"Учебник {Subject} - '{Title}' изучен");
    }
}

public class Fiction : Book
{
    public string Genre { get; set; }
    public string MainCharacter { get; set; }
    public int Rating { get; set; }
    public bool IsBestseller { get; set; }

    public Fiction(string title, string author, int yearOfPublication, string genre)
        : base(title, author, yearOfPublication)
    {
        Genre = genre;
    }

    // Перегруженный конструктор
    public Fiction(string title, string author, int yearOfPublication, 
                  string genre, string mainCharacter, int rating, bool isBestseller)
        : base(title, author, yearOfPublication)
    {
        Genre = genre;
        MainCharacter = mainCharacter;
        Rating = rating;
        IsBestseller = isBestseller;
    }

    public override void Borrow()
    {
        base.Borrow();
        if (IsBestseller)
        {
            Console.WriteLine($"Бестселлер! Рекомендуем другие книги жанра {Genre}");
        }
    }

    // Перегрузка метода Borrow
    public void Borrow(int days)
    {
        Borrow();
        Console.WriteLine($"Срок возврата: {days} дней");
    }

    // Новые методы
    public void AnalyzePlot()
    {
        Console.WriteLine($"Анализ сюжета книги {Title} с главным героем {MainCharacter}");
    }

    public void RecommendSimilar()
    {
        Console.WriteLine($"Рекомендация книг в жанре {Genre}, похожих на {Title}");
    }

    public override void Review()
    {
        Console.WriteLine($"Чтение рецензии на художественную книгу: {Title}");
        if (Rating >= 4)
        {
            Console.WriteLine("Высокий рейтинг! Стоит прочитать");
        }
    }
}

public class ScientificLiterature : Book
{
    public string FieldOfScience { get; set; }
    public string DOI { get; set; }
    public int CitationCount { get; set; }
    public bool IsPeerReviewed { get; set; }

    public ScientificLiterature(string title, string author, int yearOfPublication, string fieldOfScience)
        : base(title, author, yearOfPublication)
    {
        FieldOfScience = fieldOfScience;
    }

    // Перегруженный конструктор
    public ScientificLiterature(string title, string author, int yearOfPublication,
                               string fieldOfScience, string doi, int citationCount, bool isPeerReviewed)
        : base(title, author, yearOfPublication)
    {
        FieldOfScience = fieldOfScience;
        DOI = doi;
        CitationCount = citationCount;
        IsPeerReviewed = isPeerReviewed;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Область науки: {FieldOfScience}, Цитирований: {CitationCount}";
    }

    // Новые методы
    public void Cite()
    {
        CitationCount++;
        Console.WriteLine($"Цитирование работы: {Title}. Общее число цитирований: {CitationCount}");
    }

    public void CheckValidity()
    {
        if (IsPeerReviewed && YearOfPublication > 2010)
        {
            Console.WriteLine($"Научная работа {Title} актуальна и рецензирована");
        }
        else
        {
            Console.WriteLine($"Внимание: {Title} может требовать обновления");
        }
    }

    public override void Read()
    {
        Console.WriteLine($"Анализ научной работы по {FieldOfScience}: {Title}");
        if (IsPeerReviewed)
        {
            Console.WriteLine("Работа прошла рецензирование");
        }
    }
}

// Generic-класс библиотеки
public class Library<T> where T : Book
{
    private List<T> books = new List<T>();
    private string libraryName;

    public Library(string name)
    {
        libraryName = name;
    }

    public void AddBook(T book)
    {
        books.Add(book);
        Console.WriteLine($"Книга '{book.Title}' добавлена в библиотеку '{libraryName}'");
    }

    public void RemoveBook(T book)
    {
        if (books.Remove(book))
        {
            Console.WriteLine($"Книга '{book.Title}' удалена из библиотеки");
        }
    }

    public T FindBookByTitle(string title)
    {
        return books.Find(b => b.Title.Equals(title, StringComparison.OrdinalIgnoreCase));
    }

    public List<T> FindBooksByAuthor(string author)
    {
        return books.FindAll(b => b.Author.Equals(author, StringComparison.OrdinalIgnoreCase));
    }

    public void DisplayAllBooks()
    {
        Console.WriteLine($"\n=== Библиотека '{libraryName}' ===");
        foreach (var book in books)
        {
            Console.WriteLine(book.GetInfo());
        }
        Console.WriteLine($"Всего книг: {books.Count}\n");
    }

    // Generic-метод для обработки книг
    public void ProcessBooks<U>(Action<U> processor) where U : T
    {
        foreach (var book in books)
        {
            if (book is U specificBook)
            {
                processor(specificBook);
            }
        }
    }
}

// Дополнительный generic-класс для статистики
public class BookStatistics<T> where T : Book
{
    public int GetTotalPages(IEnumerable<T> books)
    {
        int total = 0;
        foreach (var book in books)
        {
            total += book.PageCount;
        }
        return total;
    }

    public double GetAverageYear(IEnumerable<T> books)
    {
        if (!books.Any()) return 0;
        return books.Average(b => b.YearOfPublication);
    }
}

// Создание объектов с использованием перегруженных конструкторов
Book[] books = {
    new Textbook("Алгебра", "Иванов", 2020, "Математика", 5, "10-11 класс", true),
    new Fiction("1984", "Оруэлл", 1949, "Антиутопия", "Уинстон Смит", 5, true),
    new ScientificLiterature("Квантовая физика", "Хокинг", 2015, "Физика", "10.1234/quantum", 150, true),
    new Textbook("Биология", "Петров", 2022, "Биология", 3, "9 класс", true),
    new Fiction("Мастер и Маргарита", "Булгаков", 1967, "Роман", "Мастер", 5, true)
};

// Демонстрация полиморфизма
Console.WriteLine("=== ПОЛИМОРФИЗМ ===");
foreach (var book in books)
{
    Console.WriteLine("=== Информация о книге ===");
    Console.WriteLine(book.GetInfo());
    
    Console.Write("Чтение (без параметров): ");
    book.Read();
    
    Console.Write("Чтение (с параметром): ");
    book.Read(50);
    
    Console.Write("Выдача: ");
    book.Borrow();
    
    book.Review();
    Console.WriteLine();
}

// Демонстрация перегрузки методов
Console.WriteLine("\n=== ПЕРЕГРУЗКА МЕТОДОВ ===");
Fiction fiction = (Fiction)books[1];
fiction.Borrow(30); // Перегруженный метод

Textbook textbook = (Textbook)books[0];
textbook.Read(100); // Перегруженный метод

// Демонстрация generic-класса
Console.WriteLine("\n=== GENERIC КЛАССЫ ===");

Library<Book> mainLibrary = new Library<Book>("Главная библиотека");
foreach (var book in books)
{
    mainLibrary.AddBook(book);
}

mainLibrary.DisplayAllBooks();

// Специализированная библиотека
Library<Fiction> fictionLibrary = new Library<Fiction>("Библиотека художественной литературы");
fictionLibrary.AddBook((Fiction)books[1]);
fictionLibrary.AddBook((Fiction)books[4]);
fictionLibrary.DisplayAllBooks();

// Поиск книг
var foundBook = mainLibrary.FindBookByTitle("1984");
if (foundBook != null)
{
    Console.WriteLine($"Найдена книга: {foundBook.GetInfo()}");
}

// Использование generic-метода
Console.WriteLine("\n=== ОБРАБОТКА КНИГ ===");
mainLibrary.ProcessBooks<Fiction>(f => f.AnalyzePlot());

// Использование generic-класса статистики
BookStatistics<Book> stats = new BookStatistics<Book>();
Console.WriteLine($"\nОбщее количество страниц: {stats.GetTotalPages(books)}");
Console.WriteLine($"Средний год издания: {stats.GetAverageYear(books):F0}");

// Демонстрация специфических методов
Console.WriteLine("\n=== СПЕЦИФИЧЕСКИЕ МЕТОДЫ ===");
ScientificLiterature scienceBook = (ScientificLiterature)books[2];
scienceBook.Cite();
scienceBook.CheckValidity();

textbook.DoExercises();
textbook.UpdateEdition(6);

fiction.RecommendSimilar();
```
