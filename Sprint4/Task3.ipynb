Создать базовый класс Book в C#, который будет представлять информацию о 
книгах. На основе этого класса разработать 2-3 производных класса, 
демонстрирующих принципы наследования и полиморфизма. В каждом из классов 
должны быть реализованы новые атрибуты и методы, а также переопределены 
некоторые методы базового класса для демонстрации полиморфизма.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) создайте явную реализации интерфейса и управление зависимостями 


<h2 style="color:DodgerBlue">Реализация:</h2>

```
using System;
using System.Collections.Generic;

// Интерфейс для явной реализации
public interface ILibraryItem
{
    string GetInfo();
    void Borrow();
}

// Базовый класс с новыми атрибутами и методами
public class Book
{
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public string ISBN { get; set; }           // Новый атрибут
    public decimal Price { get; set; }          // Новый атрибут
    public int PageCount { get; set; }          // Новый атрибут
    public bool IsAvailable { get; set; }       // Новый атрибут
    
    private List<string> reviews = new List<string>(); // Новый атрибут

    public Book(string title, string author, int yearOfPublication, 
                string isbn, decimal price, int pageCount)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
        ISBN = isbn;
        Price = price;
        PageCount = pageCount;
        IsAvailable = true;
    }

    public virtual string GetInfo()
    {
        return $"Название: {Title}, Автор: {Author}, Год: {YearOfPublication}, ISBN: {ISBN}";
    }

    public virtual void Read()
    {
        Console.WriteLine($"Чтение книги: {Title} ({PageCount} стр.)");
    }

    public virtual void Borrow()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            Console.WriteLine($"Выдача книги: {Title}");
        }
        else
        {
            Console.WriteLine($"Книга уже выдана: {Title}");
        }
    }

    // Новый метод
    public void Return()
    {
        IsAvailable = true;
        Console.WriteLine($"Книга возвращена: {Title}");
    }

    // Новый метод
    public void AddReview(string review)
    {
        reviews.Add(review);
        Console.WriteLine($"Отзыв добавлен к книге: {Title}");
    }

    // Новый метод
    public void DisplayReviews()
    {
        Console.WriteLine($"Отзывы о книге '{Title}':");
        foreach (var review in reviews)
        {
            Console.WriteLine($"- {review}");
        }
    }

    // Новый метод
    public virtual void CalculateReadingTime(int pagesPerDay)
    {
        int days = PageCount / pagesPerDay;
        Console.WriteLine($"На чтение книги '{Title}' потребуется {days} дней");
    }
}

public class Textbook : Book
{
    public string Subject { get; set; }
    public int GradeLevel { get; set; }         // Новый атрибут
    public bool HasExercises { get; set; }      // Новый атрибут

    public Textbook(string title, string author, int yearOfPublication, 
                   string isbn, decimal price, int pageCount,
                   string subject, int gradeLevel, bool hasExercises) 
        : base(title, author, yearOfPublication, isbn, price, pageCount)
    {
        Subject = subject;
        GradeLevel = gradeLevel;
        HasExercises = hasExercises;
    }

    public override void Read()
    {
        Console.WriteLine($"Изучение учебника '{Subject}' для {GradeLevel} класса: {Title}");
    }

    // Новый метод
    public void DoExercise(int exerciseNumber)
    {
        Console.WriteLine($"Выполнение упражнения {exerciseNumber} в учебнике: {Title}");
    }

    // Новый метод
    public void UpdateMaterial()
    {
        Console.WriteLine($"Материал учебника '{Title}' обновлён");
    }
}

public class Fiction : Book
{
    public string Genre { get; set; }
    public string MainCharacter { get; set; }   // Новый атрибут
    public bool IsBestseller { get; set; }      // Новый атрибут

    public Fiction(string title, string author, int yearOfPublication,
                  string isbn, decimal price, int pageCount,
                  string genre, string mainCharacter, bool isBestseller)
        : base(title, author, yearOfPublication, isbn, price, pageCount)
    {
        Genre = genre;
        MainCharacter = mainCharacter;
        IsBestseller = isBestseller;
    }

    public override void Borrow()
    {
        base.Borrow();
        if (IsBestseller)
        {
            Console.WriteLine($"Внимание! Это бестселлер в жанре '{Genre}'");
        }
    }

    // Новый метод
    public void AnalyzePlot()
    {
        Console.WriteLine($"Анализ сюжета. Главный герой: {MainCharacter}");
    }

    // Новый метод
    public void RateBook(int rating)
    {
        Console.WriteLine($"Книга '{Title}' оценена на {rating}/10");
    }
}

public class ScientificLiterature : Book
{
    public string FieldOfScience { get; set; }
    public int CitationCount { get; set; }      // Новый атрибут
    public string ResearchMethod { get; set; }  // Новый атрибут

    public ScientificLiterature(string title, string author, int yearOfPublication,
                               string isbn, decimal price, int pageCount,
                               string fieldOfScience, int citationCount, string researchMethod)
        : base(title, author, yearOfPublication, isbn, price, pageCount)
    {
        FieldOfScience = fieldOfScience;
        CitationCount = citationCount;
        ResearchMethod = researchMethod;
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", Область: {FieldOfScience}, Цитат: {CitationCount}";
    }

    // Новый метод
    public void Cite()
    {
        CitationCount++;
        Console.WriteLine($"Работа процитирована. Всего цитат: {CitationCount}");
    }

    // Новый метод
    public void ConductResearch()
    {
        Console.WriteLine($"Проведение исследования методом {ResearchMethod}");
    }
}

// Явная реализация интерфейса
public class AudioBook : ILibraryItem
{
    public string Title { get; set; }
    public string Author { get; set; }
    public int DurationMinutes { get; set; }
    public string Narrator { get; set; }

    string ILibraryItem.GetInfo()
    {
        return $"Аудиокнига: {Title} (исп. {Narrator}), длительность: {DurationMinutes} мин.";
    }

    void ILibraryItem.Borrow()
    {
        Console.WriteLine($"Аудиокнига выдана: {Title}");
    }

    // Собственный метод
    public void Play()
    {
        Console.WriteLine($"Воспроизведение аудиокниги: {Title}");
    }
}

// Управление зависимостями через конструктор
public class LibraryService
{
    private readonly ILogger logger;
    private List<Book> books = new List<Book>();

    // Внедрение зависимости
    public LibraryService(ILogger logger)
    {
        this.logger = logger;
    }

    public void AddBook(Book book)
    {
        books.Add(book);
        logger.Log($"Добавлена книга: {book.Title}");
    }

    public void DisplayAllBooks()
    {
        Console.WriteLine("\n=== КНИГИ В БИБЛИОТЕКЕ ===");
        foreach (var book in books)
        {
            Console.WriteLine(book.GetInfo());
        }
    }
}

// Интерфейс для зависимости
public interface ILogger
{
    void Log(string message);
}

// Реализация зависимости
public class ConsoleLogger : ILogger
{
    public void Log(string message)
    {
        Console.WriteLine($"[Лог] {DateTime.Now:HH:mm:ss}: {message}");
    }
}

// Создаем зависимости вручную
ILogger logger = new ConsoleLogger();
LibraryService library = new LibraryService(logger);

// Создаем книги с новыми атрибутами
Book[] books = {
    new Textbook("Алгебра", "Иванов", 2020, "978-123-456", 1200, 350, 
                "Математика", 10, true),
    new Fiction("1984", "Оруэлл", 1949, "978-987-654", 800, 320,
                "Антиутопия", "Уинстон Смит", true),
    new ScientificLiterature("Квантовая физика", "Хокинг", 2015, "978-111-222", 
                            2500, 500, "Физика", 150, "Экспериментальный")
};

// Добавляем книги в библиотеку
foreach (var book in books)
{
    library.AddBook(book);
}

// Демонстрация работы
foreach (var book in books)
{
    Console.WriteLine("\n" + new string('=', 40));
    Console.WriteLine(book.GetInfo());
    
    book.Read();
    book.Borrow();
    book.AddReview("Отличная книга!");
    
    // Новые методы
    book.CalculateReadingTime(50);
    
    // Методы конкретных классов
    if (book is Textbook textbook)
    {
        textbook.DoExercise(5);
        textbook.UpdateMaterial();
    }
    else if (book is Fiction fiction)
    {
        fiction.AnalyzePlot();
        fiction.RateBook(9);
    }
    else if (book is ScientificLiterature science)
    {
        science.Cite();
        science.ConductResearch();
    }
    
    book.Return();
    book.DisplayReviews();
}

// Явная реализация интерфейса
Console.WriteLine("\n" + new string('=', 40));
Console.WriteLine("ЯВНАЯ РЕАЛИЗАЦИЯ ИНТЕРФЕЙСА");

AudioBook audioBook = new AudioBook 
{ 
    Title = "Мастер и Маргарита", 
    Author = "Булгаков",
    DurationMinutes = 1420,
    Narrator = "Самойлов"
};

// Доступ через интерфейс
ILibraryItem item = audioBook;
Console.WriteLine(item.GetInfo());
item.Borrow();

// Собственные методы класса
audioBook.Play();

// Показываем все книги в библиотеке
library.DisplayAllBooks();
```
